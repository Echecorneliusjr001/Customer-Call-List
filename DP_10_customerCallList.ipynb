```python
#importing data into pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel(r"C:\Users\Cornelius\Documents\Excel hub\DP_10\DP_10_CustomerCallList.xlsx")
df
```

```python
#removing duplicates
df = df.drop_duplicates()
df
```

```python
# removing not needed column 
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Cleaning the Last_name column
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
#formatting  Phone_Number column
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-z0-9]','')

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df
```

```python
# replacing Nan & Na values in Phone_Number column
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# Standardize address column into three other columns
df[["Street_Address", "State", "Zip_code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```python
# formatting paying customer & Do_Not_Contact columns
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df ["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
df ["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df
```

```python
# fill null values 
df = df.fillna('')
df
```

```python
#Replace N/a values with empty values
df = df.replace("N/a", "")
df
```

```python
# removing values for Y in Do_Not_Contact column
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Y":
        df.drop(x, inplace=True)   
df
```

```python
#Remove empty values in Phone_Number column
for x in df.index:
    if df.loc[x, "Phone_Number"] == "":
        df.drop(x, inplace=True)   
df
```

```python
# Rearranging the index of the table
df = df.reset_index(drop=True)
df
```

```python
# Removing Zip_code column because it has only empty values
df = df.drop(columns = "Zip_code")
df
```

```python
# Removing the original Address column from the table
df = df.drop(columns = "Address")
df
```
